```python
df1 = df.groupby('Agent_ID')['Call_Status'].value_counts().unstack().reset_index()
df1.fillna(0, inplace=True)
df1['received'] = df1['answered'] + df1['transfer']
df1.drop(['answered', 'transfer'], axis = 1, inplace = True)
df1 = df1[df1['received'] == 0.0]
df1
```

```python
id_list = list(df1['Agent_ID'].values)[:-1]
name_list = [df[df['Agent_ID']==i]['Agent_Name'].values[0] for i in id_list]
# name_list
df['Agent_ID'].replace(id_list, 'not available', inplace=True)
df['Agent_Name'].replace(name_list, 'not available', inplace=True)
df.head(10)
```
